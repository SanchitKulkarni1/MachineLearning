```python
#Training our datasets: 80% will be used for training and 20% for testing
msk=np.random.rand((len(df))) <0.8
train=df[msk]
test=df[~msk]
```

```python
#Implementing Simple Linear Regression
#Train Data Distribution

plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
#Modeling: using sklearn package to model data

from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit(train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
#Evaluation
from sklearn.metrics import r2_score

test_x=np.asanyarray(test[['ENGINESIZE']])
test_y=np.asanyarray(test[['CO2EMISSIONS']])
test_y_hat=regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat-test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat-test_y)**2))
print("R2-score: %.2f" % r2_score(test_y_hat,test_y))
```

```python
train_x= train[["FUELCONSUMPTION_COMB"]]
test_x= test[["FUELCONSUMPTION_COMB"]]
```

```python
regr=linear_model.LinearRegression()
regr.fit(train_x,train_y)
```

```python
prediction=regr.predict(test_x)
```

```python
print("Mean absolute error: %.2f" %np.mean(np.absolute(prediction - test_y)))
print("Residual sum of squares (MSE): %.2f"% np.mean(prediction - test_y))
```

```python
#Implementing the trained model on the new data set : Using Engine Size as the independent variable
new_data = {'ENGINESIZE': [2.0, 2.4, 3.5, 1.5, 3.0]}
new_data_df = pd.DataFrame(new_data)
new_data_df.describe()
```

```python
new_x=np.asanyarray(new_data_df[['ENGINESIZE']])
predictions=regr.predict(new_x)
new_data['PREDICTED_CO2']=predictions
print(new_data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(new_data_df.ENGINESIZE, new_data['PREDICTED_CO2'], color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
